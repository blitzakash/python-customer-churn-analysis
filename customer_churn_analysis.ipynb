# Customer Churn Analysis using Python

This project analyzes customer churn data to understand patterns in customer behavior and identify factors that influence churn.

The analysis includes:
- Data exploration
- Data cleaning
- Customer churn analysis
- Visualizations
- Business insights

Tools used:
- Python
- Pandas
- Matplotlib
- Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data["Churn"].value_counts()
```

```python
sns.countplot(x="Churn", data=data)

plt.title("Customer Churn Distribution")
plt.show()
```

```python
sns.countplot(x="Contract", hue="Churn", data=data)

plt.title("Churn by Contract Type")
plt.show()
```

```python
sns.boxplot(x="Churn", y="MonthlyCharges", data=data)

plt.title("Monthly Charges vs Churn")
plt.show()
```

```python
sns.boxplot(x="Churn", y="tenure", data=data)

plt.title("Customer Tenure vs Churn")
plt.show()
```

## Key Insights

From the churn analysis, several important patterns were observed:

1. Customers with **month-to-month contracts** have the highest churn rate.
2. Customers with **higher monthly charges** are more likely to churn.
3. Customers with **short tenure** are more likely to leave the service.
4. Long-term contracts (one-year or two-year) significantly reduce churn.

These insights suggest that companies should focus on improving retention strategies for new customers and customers with month-to-month contracts.
